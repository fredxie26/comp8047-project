```python
import pyodbc

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=;'
                      'Database=comp8047;'
                      'Trusted_Connection=yes;')
```

```python
'''
top company by layoff numbers
'''
df_1 = pd.read_sql("select top 20 company, sum(total_laid_off) as total_laidoff from layoff_data group by company order by total_laidoff desc", conn)

fig = plt.subplots(figsize=(16, 5))
plt.bar(df_1['company'], df_1['total_laidoff'], align='center', alpha=0.5)
plt.ylabel('total laid off')
plt.title('company')
plt.xticks(rotation=45);
plt.show()
```

```python
'''
top company by layoff ratio
'''
df_2 = pd.read_sql("select top 20 company, sum(percentage_laid_off) as total_laidoff from layoff_data group by company order by company", conn)

fig = plt.subplots(figsize=(16, 5))
plt.bar(df_2['company'], df_2['total_laidoff'], align='center', alpha=0.5)
plt.ylabel('total laid off ratio')
plt.title('company')
plt.xticks(rotation=45);
plt.show()
```

```python
'''
top company by layoff numbers funds raised
'''
df_3 = pd.read_sql("with l as (select top 20 company, sum(percentage_laid_off) as total_laidoff from layoff_data group by company order by company) " \
                   "select l.company, sum(f.funds_raised) as total_funds_raised " \
                   "from l join layoff_data f on l.company = f.company group by l.company order by l.company", conn)

fig = plt.subplots(figsize=(16, 5))
plt.bar(df_3['company'], df_3['total_funds_raised'], align='center', alpha=0.5)
plt.ylabel('total funds raised')
plt.title('company')
plt.xticks(rotation=45);
plt.show()
```
